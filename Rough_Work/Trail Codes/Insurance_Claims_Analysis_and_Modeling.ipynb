```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Simulating sample data
np.random.seed(42)
data = pd.DataFrame({
    'Credit Score': np.random.normal(650, 50, 1000),
    'Claim Amount': np.random.exponential(2000, 1000),
    'Age': np.random.randint(18, 80, 1000),
    'Past Claims': np.random.randint(0, 5, 1000),
    'Future Claim': np.random.choice([0, 1], size=1000, p=[0.7, 0.3])
})
```

```python
# 1. Claim Distribution Bar Chart
plt.figure(figsize=(6, 4))
sns.countplot(x=data["Future Claim"])
plt.title("Claim Distribution (Future Claims vs. No Claims)")
plt.xlabel("Claim Status")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["No Claim", "Claim"])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("claim_distribution.png")
plt.show()
```

```python
# 2. Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(data[['Credit Score', 'Claim Amount', 'Age', 'Past Claims', 'Future Claim']].corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.savefig("feature_correlation.png")
plt.show()
```

```python
# 3. Model Comparison Bar Chart
models = ['KNN', 'Decision Tree', 'SVM', 'Logistic Regression']
jaccard_scores = [0.72, 0.75, 0.78, 0.82]
f1_scores = [0.71, 0.74, 0.77, 0.81]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(x - width/2, jaccard_scores, width, label='Jaccard Score', color='b', alpha=0.7)
ax.bar(x + width/2, f1_scores, width, label='F1 Score', color='r', alpha=0.7)
ax.set_ylabel('Score')
ax.set_title('Model Comparison (Jaccard & F1 Score)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("model_comparison.png")
plt.show()
```

```python
# 4. Feature Importance Chart
features = ["Credit Score", "Age", "Past Claims", "Claim Amount", "Vehicle Ownership", "Marital Status"]
importance = np.random.rand(len(features)) * 100

plt.figure(figsize=(6, 4))
sns.barplot(x=importance, y=features, palette="Blues_r")
plt.title("Feature Importance (Best Model)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.savefig("feature_importance.png")
plt.show()
```
