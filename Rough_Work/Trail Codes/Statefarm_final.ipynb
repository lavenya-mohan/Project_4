```python
import pandas as pd
import numpy as np
```

```python
cd = pd.read_excel('/content/Claim_Data.xlsm')
```

```python
cd.head(10)
```

```python
cd.dtypes
```

```python
cd_new = cd.loc[(cd['clm_dates'] >= '2012-1-1') & (cd['clm_dates'] < '2017-1-1')]
cd_new
```

```python
grouped = cd_new.groupby(cd_new.aft_ind)
cd_1 = grouped.get_group(0)
cd_1
```

```python
A = cd_1.replace(to_replace =0,
                 value =1)
```

```python
df = A.rename(columns={'aft_ind':'ct_not_fault'}, inplace=False)
```

```python
df['hhld_id'].is_unique
```

```python
cols = ['hhld_id','ct_not_fault']
```

```python
df = df[cols]
df.columns
```

```python
df.dtypes
```

```python
df_agg = df.groupby(['hhld_id'], as_index=False)['ct_not_fault'].sum()
df_agg
```

```python
df_agg.describe()
```

```python
cd_2 = grouped.get_group(1)
cd_2
```

```python
df_1 = cd_2.rename(columns={'aft_ind':'ct_at_fault'}, inplace=False)
```

```python
cols_1 = ['hhld_id','ct_at_fault']
```

```python
df_1 = df_1[cols_1]
df_1.columns
```

```python
df_agg_1 = df_1.groupby(['hhld_id'], as_index=False)['ct_at_fault'].sum()
df_agg_1
```

```python
result = pd.merge(df_agg, df_agg_1[['hhld_id', 'ct_at_fault']], on ='hhld_id', how='outer')
result
```

```python
result['hhld_id'].is_unique
```

```python
result.replace(np.NaN, 0)
```

```python
claim_data=result.replace(np.NaN, 0)
claim_data
```

```python
# Converting 'ct_not_fault, ct_at_fault' columns from float to int datatype
claim_data['ct_not_fault'] = claim_data['ct_not_fault'].apply(np.int64)
claim_data['ct_at_fault'] = claim_data['ct_at_fault'].apply(np.int64)
claim_data
```

```python
claim_data.describe()


```

```python
#Load Predictor dataset
pd_df=pd.read_excel('/content/Predictor Dataset.xlsm')
pd_df.head()
```

```python
#Load Subsequent loss dataset
sl_df=pd.read_excel('/content/Subsequent_Loss_Experience.xlsm')
sl_df.head()
```

```python
#Merging  Predictor & Subsequent Loss datasets with aggregated Claim dataset to create Model
all_df = [pd_df, claim_data, sl_df]
all_df
```

```python
#Use reduce funcation to merge all 3 datasets 
#Merging  Predictor & Subsequent Loss datasets with aggregated Claim dataset to create Model
from functools import reduce
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['hhld_id'],
                                            how='outer'), all_df)
```
